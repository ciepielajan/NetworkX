```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
```

```python
g.add_node(1,pos=(0,6))
g.add_node(2,pos=(8,0))
g.add_node(3,pos=(11,4))
g.add_node(4,pos=(6,11))
g.add_node(5,pos=(15,0))
g.add_node(6,pos=(16,6))
g.add_node(7,pos=(17,10))
g.add_node(8,pos=(10,15))
g.add_node(9,pos=(16,13))
g.add_node(10,pos=(20,80))
```

```python
pos = nx.spring_layout(g, seed=3068)  # Seed layout for reproducibility
nx.draw(g, pos=pos, with_labels=True)
plt.show()
```

```python
g.add_edge(1,4,weight=900)
g.add_edge(4,8,weight=319)
g.add_edge(4,7,weight=418)
g.add_edge(8,9,weight=200)
g.add_edge(9,7,weight=30)
g.add_edge(7,10,weight=50)
g.add_edge(4,6,weight=210)
g.add_edge(6,10,weight=698)
g.add_edge(1,3,weight=510)
g.add_edge(1,2,weight=320)
g.add_edge(2,3,weight=176)
g.add_edge(3,6,weight=290)
g.add_edge(3,5,weight=350)
g.add_edge(5,6,weight=613)
```

```python
nx.draw(g,pos)
labels = nx.get_edge_attributes(g,'weight')
nx.draw_networkx_edge_labels(g,pos,edge_labels=labels)
plt.show()

# jak dodać nazwy punktów
# jak dodać km do labeles
# jak dodać kierunek strzałek 
# jak zmienić rozmiar grafu coś jak figsize
# jak generowac zawsze takie same grafy np  jakaś zmienna jak  seed
# jak wygenerowac mozliwe wszystkie drogi 
# szybkie wczytanie przykładowego dużego grafu z weight
# czy poza weight mozna coś jeszcze dodać (jeżeli tak to czy musi być w kazdym połączeniu )
# czy da się zamrozić jakieś połącznie np miedzy 9 a 10
```
